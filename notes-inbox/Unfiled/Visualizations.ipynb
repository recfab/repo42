## Normal Distribution

Select items from a normal, random distribution

```python
SIZE=10000
x = np.random.normal(size=SIZE)
x
```

When displayed as a histogram, there will be a peak in the center i.e. a bell curve

```python
plt.hist(x, bins=50)
```

## Uniform Distribution

```python
y = np.random.normal(size=SIZE)
yu = np.random.uniform(size=SIZE)
```

```python
plt.scatter(x, y)
```

```python
xu = np.random.uniform(size=SIZE)
plt.scatter(xu, yu)
```

# Seaborn

<http://seaborn.pydata.org>

`matplotlib` can be hard to use. Seaborn is built on top osf it and is easier to use.

```python
import seaborn as sns
sns.set_theme(style="ticks")

# Load the example dataset for Anscombe's quartet
df = sns.load_dataset("anscombe")

# Show the results of a linear regression within each dataset
sns.lmplot(x="x", y="y", col="dataset", hue="dataset", data=df,
           col_wrap=2, ci=None, palette="muted", height=4,
           scatter_kws={"s": 50, "alpha": 1})
```
