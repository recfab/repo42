# Sieve of Aristophanes

Generates prime numbers up to a given limit

```python
def sieve(n):
  """
  Generates prime numbers up to n
  """
  not_prime = []
  for i in range(2, n + 1):
    if i not in not_prime:
      print(i)
      for j in range(i * i, n + 1, i):
        not_prime.append(j)
```
