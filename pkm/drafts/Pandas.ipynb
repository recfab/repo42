# Pandas

- Python library for data analysis
- "mimics a lot of the functionality of R; so you get the power of R, without the quirks or R."[^1]
- 2D data structure: "Data Frame"
  - can load data into the Frame e.g. using `pandas.read_csv()`
  - display in Jupyter with rich output
  - styled using HTML
  - can access styles using `data_frame.style` property and its member functions
